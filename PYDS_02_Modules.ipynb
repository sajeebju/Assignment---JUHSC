### Group Indices

```python
# Index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index
```

```python
hier_index = pd.MultiIndex.from_tuples(hier_index, names=('main', 'sub'))
hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[2]
```

```python
df.index.names
```

```python
# a different way
df.xs('G1')
```

```python
df.xs(2,level='sub')
```
