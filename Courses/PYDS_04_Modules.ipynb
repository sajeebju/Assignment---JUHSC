```python
import numpy as np
x = np.linspace(1, 10, 12)
y1 = x**2
y2 = x**3
y3 = x**3-x**2+x-5
y4 = np.sin(x)
y5 = np.cos(x)



# just to check the difference between math module 
# functions and numpy module functions
import math
y7 = math.sin(45)

print(x)
print(y1)
print(y7)
```

## Basic Plotting

```python
plt.plot(x, y3, 'black')                  # 'g' is a color code
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('Speed with time')
plt.show()
```

## Plotting using OOP

```python
# Basic OOP Concept

class Human():
    def __init__(self, name, height):
        self.name = name
        self.height = height
        
    def greetings(self):
        print("Hello", self.name)
        
    def printyourname(self):
        print("Your name is:", self.name)
        
    def heightconvert(self):
        print("your height is:", int(self.height)*12) # wrong idea
        
        
a = Human("Roy", 5.7)
a.greetings()
a.printyourname()

b = Human("Rishal", 6)
b.greetings()

c = Human("Shahriar", 5.8)
c.greetings()
c.heightconvert()
```

```python
# Another Example
class Subject():
    
    def __init__(self, name, mark):
        self.name = name
        self.mark = mark
        
    def grade_point(self):
        if 80<=self.mark<=100:
            return 4
        elif 60<=self.mark<=79:
            return 3
        elif 40<=self.mark<=59:
            return 2
        else:
            return 0
    
    def grade(self):
        if 80<=self.mark<=100:
            return "A"
        elif 60<=self.mark<=79:
            return "B"
        elif 40<=self.mark<=59:
            return "C"
        else:
            return 0
        
        
bangla = Subject("Bangla", 74)
print(bangla.grade_point())
print(bangla.grade())
```

```python
# creating an object of plt.figure() class
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1]) # x0, y0, width, height (range 0 to 1)
# axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 

# Plot on that set of axes
axes.plot(x, y4, 'bo-')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Example Title');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # main axes
axes2 = fig.add_axes([0.65, 0.25, 0.25, 0.2]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y4, 'b*-')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(x, y5, 'r+-')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Axes 2 Title');
```
